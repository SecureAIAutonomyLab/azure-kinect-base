```python
import asyncio
import io
import glob
import os
import sys
import time
import uuid
import requests
from urllib.parse import urlparse
from io import BytesIO
from PIL import Image, ImageDraw
from azure.cognitiveservices.vision.face import FaceClient
from msrest.authentication import CognitiveServicesCredentials
from azure.cognitiveservices.vision.face.models import TrainingStatusType, Person, SnapshotObjectType, OperationStatusType
import csv
import errno
import json
# Set the FACE_SUBSCRIPTION_KEY environment variable with your key as the value.
# This key will serve all examples in this document.
KEY = 'your key here'

# Set the FACE_ENDPOINT environment variable with the endpoint from your Face service in Azure.
# This endpoint will be used in all examples in this quickstart.
ENDPOINT = 'your endpoint here'
# Create an authenticated FaceClient.
face_client = FaceClient(ENDPOINT, CognitiveServicesCredentials(KEY))
```

```python
path = './rgbimgs/1-23-120/14:14:44/*.png'   
files = glob.glob(path)
headers = ["File_Name","Date","Time","File_Type","Face_ID", "Name","Face_Rectangle_Top","Face_Rectangle_Left",
              "Face_Rectangle_Width","Face_Rectangle_Height","Face_Attributes_Smile","HeadPose_Pitch","HeadPose_Roll",
              "HeadPose_yaw","Gender","Age","Facial_Hair_Moustache","Facial_Hair_Beard","Facial_Hair_Sideburns",
              "Glasses","Anger","Contempt","Disgust","Fear","Happiness","Neutral","Sadness","Surprise","Blur_Level","Blur_Value",
              "Exposure_Level","Exposure_Value","Noise_Level","Noise_Value","Eye_Makeup","Lip_Makeup","Accessories","Forehead_Occluded",
              "Eye_Occluded","Mouth_Occluded","Hair_Bald","Hair_Invisible","Hair_Color_Brown_Confidence","Hair_Color_Black_Confidence",
              "Hair_Color_Gray_Confidence","Hair_Color_Blond_Confidence","Hair_Color_Red_Confidence","Hair_Color_Other_Confidence"]

#face_client.person_group.create(person_group_id=PERSON_GROUP_ID, name=PERSON_GROUP_ID)
PERSON_GROUP_ID = 'lab-group'
# Define person friend
josh = face_client.person_group_person.create(PERSON_GROUP_ID, "josh")
# Define man friend
avery = face_client.person_group_person.create(PERSON_GROUP_ID, "avery")

josh_images = [file for file in glob.glob('*.jpg') if file.startswith("josh")]
for image in josh_images:
    a = open(image, 'r+b')
    face_client.person_group_person.add_face_from_stream(PERSON_GROUP_ID, josh.person_id, a)
    
avery_images = [file for file in glob.glob('*.jpg') if file.startswith("avery")]
for image in avery_images:
    a = open(image, 'r+b')
    face_client.person_group_person.add_face_from_stream(PERSON_GROUP_ID, avery.person_id, a)

# Add to a child person
#for image in child_images:
    #ch = open(image, 'r+b')
    #face_client.person_group_person.add_face_from_stream(PERSON_GROUP_ID, child.person_id, ch)

    '''
Train PersonGroup
'''
print()
print('Training the person group...')
# Train the person group
face_client.person_group.train(PERSON_GROUP_ID)
print(josh.person_id)
print(avery.person_id)
while (True):
    training_status = face_client.person_group.get_training_status(PERSON_GROUP_ID)
    print("Training status: {}.".format(training_status.status))
    print()
    if (training_status.status is TrainingStatusType.succeeded):
        break
    elif (training_status.status is TrainingStatusType.failed):
        sys.exit('Training the person group has failed.')
        

person_ids = [josh.person_id,avery.person_id]
person_ids_dict = {josh.person_id:"joshua",avery.person_id:"avery"}

with open('face_detect.csv', mode='w') as face_detect:
    face_writer = csv.writer(face_detect, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    face_writer.writerow(headers)
    for name in files:
        img_name = list(name)
        for i in range(0,28):
            img_name.remove(img_name[0])
        empty_space = ""
        img_name = empty_space.join(img_name)
        single_face_image_url = 'https://azure-face-detect.s3.amazonaws.com/'+ img_name
        single_image_name = os.path.basename(single_face_image_url)
        detected_faces = face_client.face.detect_with_url(url=single_face_image_url)
        if not detected_faces:
            print('No face detected from image {}'.format(single_image_name)) 
        # Convert width height to a point in a rectangle
        def getRectangle(faceDictionary):
            rect = faceDictionary.face_rectangle
            left = rect.left
            top = rect.top
            bottom = left + rect.height
            right = top + rect.width
            return ((left, top), (bottom, right))


        # Download the image from the url
        response = requests.get(single_face_image_url)
        img = Image.open(BytesIO(response.content))

        # For each face returned use the face rectangle and draw a red box.
        print('Drawing rectangle around face... see popup for results.')
        draw = ImageDraw.Draw(img)
        for face in detected_faces:
            draw.rectangle(getRectangle(face), outline='red')

        # Display the image in the users default image browser.
        #img.show()



        # replace <My Endpoint String> with the string from your endpoint URL
        face_api_url = 'https://azurefacetest.cognitiveservices.azure.com/face/v1.0/detect'

        headers = {'Ocp-Apim-Subscription-Key': KEY}

        params = {
            'returnFaceId': 'true',
            'returnFaceLandmarks': 'false',
            'returnFaceAttributes': 'age,gender,headPose,smile,facialHair,glasses,emotion,hair,makeup,occlusion,accessories,blur,exposure,noise',
        }

        response = requests.post(face_api_url, params=params,
                            headers=headers, json={"url": single_face_image_url})

       

        api_output = (json.dumps(response.json()))
        api_dictionary = json.loads(api_output)

        for human in api_dictionary:
            values = []
            file_name = img_name
            Date = "2/19/20"
            Time = "modelo time"
            File_Type = "png"
            values.append(file_name)
            values.append(Date)
            values.append(Time)
            values.append(File_Type)
            values.append(human["faceId"])
            for person in person_ids:
                results = face_client.face.verify_face_to_person(human["faceId"], person, person_group_id=PERSON_GROUP_ID)
                if results.is_identical == True:
                    values.append(person_ids_dict[person])
                    val = 1
                    break
                else:
                    val = 0
            if val == 0:
                values.append("unknown")    
            values.append(human["faceRectangle"]["top"])
            values.append(human["faceRectangle"]["left"])
            values.append(human["faceRectangle"]["width"])
            values.append(human["faceRectangle"]["height"])
            values.append(human["faceAttributes"]["smile"])
            values.append(human["faceAttributes"]["headPose"]["pitch"])
            values.append(human["faceAttributes"]["headPose"]["roll"])
            values.append(human["faceAttributes"]["headPose"]["yaw"])
            values.append(human["faceAttributes"]["gender"])
            values.append(human["faceAttributes"]["age"])
            values.append(human["faceAttributes"]["facialHair"]["moustache"])
            values.append(human["faceAttributes"]["facialHair"]["beard"])
            values.append(human["faceAttributes"]["facialHair"]["sideburns"])
            values.append(human["faceAttributes"]["glasses"])
            values.append(human["faceAttributes"]["emotion"]["anger"])
            values.append(human["faceAttributes"]["emotion"]["contempt"])
            values.append(human["faceAttributes"]["emotion"]["disgust"])
            values.append(human["faceAttributes"]["emotion"]["fear"])
            values.append(human["faceAttributes"]["emotion"]["happiness"])
            values.append(human["faceAttributes"]["emotion"]["neutral"])
            values.append(human["faceAttributes"]["emotion"]["sadness"])
            values.append(human["faceAttributes"]["emotion"]["surprise"])
            values.append(human["faceAttributes"]["blur"]["blurLevel"])
            values.append(human["faceAttributes"]["blur"]["value"])
            values.append(human["faceAttributes"]["exposure"]["exposureLevel"])
            values.append(human["faceAttributes"]["exposure"]["value"])
            values.append(human["faceAttributes"]["noise"]["noiseLevel"])
            values.append(human["faceAttributes"]["noise"]["value"])
            values.append(human["faceAttributes"]["makeup"]["eyeMakeup"])
            values.append(human["faceAttributes"]["makeup"]["lipMakeup"])
            values.append(human["faceAttributes"]["accessories"])
            values.append(human["faceAttributes"]["occlusion"]["foreheadOccluded"])
            values.append(human["faceAttributes"]["occlusion"]["eyeOccluded"])
            values.append(human["faceAttributes"]["occlusion"]["mouthOccluded"])
            values.append(human["faceAttributes"]["hair"]["bald"])
            values.append(human["faceAttributes"]["hair"]["invisible"])
            values.append(human["faceAttributes"]["hair"]["hairColor"][0]["confidence"])
            values.append(human["faceAttributes"]["hair"]["hairColor"][1]["confidence"])
            values.append(human["faceAttributes"]["hair"]["hairColor"][2]["confidence"])
            values.append(human["faceAttributes"]["hair"]["hairColor"][3]["confidence"])
            values.append(human["faceAttributes"]["hair"]["hairColor"][4]["confidence"])
            values.append(human["faceAttributes"]["hair"]["hairColor"][5]["confidence"])
            face_writer.writerow(values)

```
